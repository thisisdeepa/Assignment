```python
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
df.head(5)
```

```python
df.describe()
```

```python
print(df["cast"])
```

```python
df_new = df.dropna(subset=['cast'])

actor_data = df_new['cast'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).reset_index()
actor_data.columns = ['Title_Index', 'Actor_name']

print(actor_data.head(20))
```

```python
actor_freq = actor_data['Actor_name'].value_counts().reset_index()
actor_freq.columns = ['Actor_name', 'Frequency']
print("Actor names with their frequency of appearance\n")
print(actor_freq)
```

```python
top_10_actors = actor_freq.head(10)
print("Top 10 Actors with their frequency of appearance\n")
print(top_10_actors) 
```
